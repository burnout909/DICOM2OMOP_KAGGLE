```python
import pandas as pd

valuesets = pd.read_csv("./files/part16_fhir_valuesets.csv")
valuesets.head()
```

```python
# Function to extract the numeric value after "cid-"
def extract_cid_value(text):
    start_index = text.find('cid-') + 4
    end_index = text.find('-', start_index)
    return text[start_index:end_index]

valuesets['cid'] = valuesets['id'].apply(extract_cid_value)
```

```python
print(valuesets.system.unique().size)
print(valuesets.system.unique())
```

## Inspect SNOMED and LOINC

```python
valuesets_snomed = valuesets[valuesets['system']=='http://snomed.info/sct']
valuesets_snomed #16873, code
```

```python
valuesets_loinc = valuesets[valuesets['system']=='http://loinc.org']
valuesets_loinc #2257, code
```

```python
import csv
import pandas as pd

# Open the CSV file using the csv module
with open('./OMOP_vocabulary_v5/CONCEPT.csv', 'r', newline='') as file:
    # Use a CSV reader with custom delimiter and quoting to handle commas within a column
    reader = csv.reader(file, delimiter='\t', quotechar='"')
    
    # Read the CSV data into a list of rows
    data = list(reader)

# Determine the number of columns based on the first row
num_columns = len(data[0])

# Filter out rows with incorrect number of columns
data_filtered = [row for row in data if len(row) == num_columns]

# Convert the list of rows into a DataFrame
concept = pd.DataFrame(data_filtered[1:], columns=data_filtered[0])
```

```python
concept
```

```python
concept_snomed = concept[concept['vocabulary_id']=='SNOMED']
concept_loinc = concept[concept['vocabulary_id']=='LOINC']
```

```python
concept_snomed #concept_code
```

```python
combined_snomed = pd.merge(valuesets_snomed, concept_snomed, left_on = 'code', right_on = 'concept_code', how = 'left')
```

```python
missing_snomed = combined_snomed[combined_snomed['concept_id'].isna()]
print(missing_snomed.shape[0], round(missing_snomed.shape[0]/combined_snomed.shape[0]*100,2), '%')
```

```python
combined_loinc = pd.merge(valuesets_loinc, concept_loinc, left_on = 'code', right_on = 'concept_code', how = 'left')
```

```python
missing_loinc = combined_loinc[combined_loinc['concept_id'].isna()] 
print(missing_loinc.shape[0], round(missing_loinc.shape[0]/combined_loinc.shape[0]*100,2), '%')
```

```python
print('SNOMED Value Set:', missing_snomed.version.max(), 'OMOP Concept:', concept_snomed.valid_start_date.max())
print('LOINC Value Set:', missing_loinc.version.max(), 'OMOP Concept:', concept_loinc.valid_start_date.max())
# for SNOMED codes, ask why they are not up to date to OMOP CDM vocab team; it might be coming in the next release.
```
